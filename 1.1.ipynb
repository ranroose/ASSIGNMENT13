```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics

# Data from previous assignment
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct']
income = np.array([10000, 15000, 12000, 18000, 17000, 16000, 19000, 22000, 21000, 25000])
income_after_tax = income * 0.7
expenses = np.array([7000, 8000, 6500, 9000, 8500, 8000, 9500, 10000, 9500, 12000])
savings = income_after_tax - expenses

# Create DataFrame
data = {'Income (After Tax)': income_after_tax, 'Expenses': expenses, 'Savings': savings}
df = pd.DataFrame(data, index=months)

print("Company Financial Data (10 months):")
print(df)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(months, income_after_tax, marker='o', linestyle='-', color='blue')
plt.title('Income by Month')
plt.xlabel('Month')
plt.ylabel('Income (After Tax)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(months, savings, color='green')
plt.title('Savings by Month')
plt.xlabel('Month')
plt.ylabel('Savings')
plt.show()
```

```python
colors = plt.cm.Paired(np.linspace(0, 1, len(months)))
plt.figure(figsize=(8, 8))
plt.pie(savings, labels=months, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Savings Contribution by Month')
plt.axis('equal')
plt.show()
```

```python
quarters = {
    'Q1': ['Jan', 'Feb', 'Mar'],
    'Q2': ['Apr', 'May', 'Jun'],
    'Q3': ['Jul', 'Aug', 'Sep'],
    'Q4': ['Oct']
}

for q, months_in_q in quarters.items():
    incomes = df.loc[months_in_q, 'Income (After Tax)']
    avg_income = statistics.mean(incomes)
    print(f'Average Income for {q}: {avg_income:.2f}')
```
